```python
import numpy as np
import matplotlib.pyplot as plt

initial_investment = 1000
days = 252
mu = 0.001  # expected daily return
sigma = 0.02  # daily volatility
simulations = 1000

# our simulation 
all_paths = []
for _ in range(simulations):
    daily_returns = np.random.normal(mu, sigma, days) # we are simulating using a normal distribution mu 
    price_path = initial_investment * np.cumprod(1 + daily_returns)
    all_paths.append(price_path)

all_paths = np.array(all_paths) #convert to array for easier plotting 

expected_path = np.mean(all_paths, axis=0) #calculate the expected value at each day

# plot the simulation
plt.figure(figsize=(10, 6))
for i in range(100): #we are only doing 100 so our graph isn't cluttered
    plt.plot(all_paths[i], linewidth=0.5)

# plot the expected path
plt.plot(expected_path, color='black', linewidth=3, label='Expected Portfolio Value')

plt.title("Monte Carlo Simulation of Portfolio Value Over 1 Year")
plt.xlabel("Days")
plt.ylabel("Portfolio Value ($)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
